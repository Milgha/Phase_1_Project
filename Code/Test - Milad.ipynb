```python
sql_final_table = pd.read_sql('''
SELECT primary_title, start_year, runtime_minutes, genres, averagerating, numvotes, primary_name, category
FROM principals
JOIN persons
    USING(person_id)
JOIN movie_basics
    USING(movie_id)
JOIN movie_ratings
    USING(movie_id)
WHERE start_year > 2017
AND numvotes > 100
''', con)

sql_final_table
```
